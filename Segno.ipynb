```python
import io
import segno
from PIL import Image
```

```python
qrcode = segno.make("Quick video on Segno")
qrcode.save("qr_code.png",scale=5)
Image.open("qr_code.png")
```

```python
qr_video = segno.make("https://www.youtube.com/watch?v=-9JTgHdcLCI")
qr_video.save("link.png", scale=6, dark='blue', light='black')
Image.open("link.png")
```

```python
from segno import helpers

wifi_setting = {
    'ssid': 'mywifi',
    'password': '123455678',
    'security': 'WPA'
}

wifi = helpers.make_wifi(**wifi_setting)
wifi.save("wifi.png", dark='orange',scale=6)
Image.open("wifi.png")
```

```python
qr_code = segno.make('qrcode saved in buffer memory')

out = io.BytesIO()
qr_code.save(out, scale=5, kind='png', dark='purple')

out.seek(0)
Image.open(out)
```

```python
qr_logo = segno.make("https://www.youtube.com/watch?v=-9JTgHdcLCI", error='h')
qr_logo.save('qr_logo.png', scale=8)

img = Image.open('qr_logo.png')
img = img.convert('RGB')

img_width, img_height = img.size
logo_img = Image.open('p1.png')

logo_max_size = img_height // 3

logo_img.thumbnail((logo_max_size, logo_max_size), Image.LANCZOS)
box = ((img_width - logo_img.size[0]) // 2, (img_height - logo_img.size[1]) // 2)

img.paste(logo_img, box)
img.save('qrcode_with_logo.png')
Image.open('qrcode_with_logo.png')
```
